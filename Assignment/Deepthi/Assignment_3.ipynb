# Split Dependent and independent variables

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Age',axis = 1)
y = df['Age']
```

# Scale the Independent Dataset

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
```

```python
Scale_x = pd.DataFrame(Scaler.fit_transform(x))
```

```python
Scale_x.head()
```

# Split into Training and Testing Data

```python
x_train,x_test,y_train,y_test = train_test_split(Scale_x,y,random_state = 100, train_size = 0.75)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# Build The Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
Model1 = LinearRegression()
Model1.fit(x_train,y_train)
```

```python
Pred1 = Model1.predict(x_test)
```

```python
Pred1
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print(r2_score(y_test,Pred1))
print(mean_squared_error(Pred1,y_test))
```

```python
Model2 = DecisionTreeRegressor()
Model2.fit(x_train,y_train)
```

```python
Pred2 = Model2.predict(x_test)
```

```python
Pred2
```

```python
print(r2_score(y_test,Pred2))
print(mean_squared_error(Pred2,y_test))
```

```python
Model3 = RandomForestRegressor()
Model3.fit(x_train,y_train)
```

```python
Pred3 = Model3.predict(x_test)
```

```python
Pred3
```

```python
print(r2_score(y_test,Pred3))
print(mean_squared_error(Pred3,y_test))
```

```python
from sklearn.linear_model import Lasso
```

```python
Model4 = Lasso()
Model4.fit(x_train,y_train)
```

```python
Pred4 = Model4.predict(x_test)
```

```python
Pred4
```

```python
print(r2_score(y_test,Pred4))
print(mean_squared_error(Pred4,y_test))
```
